```python
import os
import subprocess
```

```python
new_root = "../../midi/lmd_matched_flat_musicxml_test"
abc_root = "../../midi/lmd_matched_flat_abc_test_from_xml2"
```

```python
os.makedirs(abc_root)
```

```python
wrong_files = []
```

```python
for root, folders, files in os.walk(new_root):
    for file in files:
        midi_path = os.path.join(root, file)
        abc_folder = os.path.join(abc_root, root.replace(new_root + "/", ""))
        #abc_path = os.path.join(abc_folder, file.replace(".musicxml", ".abc").replace(".mid", ".abc"))
        
        # L:1/1 | -K:C | one bar per line | no group | no lyrics
        command = "xml2abc.exe {} -o {}".format(midi_path, abc_folder)

        os.makedirs(abc_folder, exist_ok=True)
        
        try:
            subprocess.run(command, shell=True, check=True)
        except subprocess.CalledProcessError as e:
            wrong_files.append(midi_path)
            print(f"Error executing command on {midi_path}: {e}")
```

```python
import json
with open("wrongs.json", "w") as wrongs_file:
    json.dump(wrong_files, wrongs_file)
```
