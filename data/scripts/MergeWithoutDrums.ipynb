# Code

```python
import mido
import os
```

```python
new_root = "lmd_matched_flat"
```

```python
merged_new_root = "lmd_matched_flat_merged_without_drums"
```

```python
os.makedirs(merged_new_root)
```

```python
wrong_files = []
```

```python
def merge_channels_without_drums(mid):
    #merged_track = mido.MidiTrack()
    for track in mid.tracks:
        for msg in track:
            if hasattr(msg, "channel") and msg.channel != 9 and msg.type != 'program_change':
                msg.channel = 0
    return mid
```

```python
#mid = mido.MidiFile("39d6c288e1bd93d4705e7a86555347da.mid")
#merge_channels_without_drums(mid)
```

```python
def gen_merged_mid(input_filename, output_filename):
    try:
        mid = mido.MidiFile(input_filename)        
        merged_mid = merge_channels_without_drums(mid)
        
        #new_mid = mido.MidiFile()
        #new_mid.tracks.append(merged_track)
        merged_mid.save(output_filename)
    except Exception as e:
        wrong_files.append(input_filename)
        print(f"Error executing on {input_filename}: {e}")
```

```python
for root, folders, files in os.walk(new_root):
    if len(files) == 0:
        # print("[NOTHING] root: {} | folders: {} | files: {}".format(root, folders, files))
        continue

    last_folder = os.path.basename(root)
    new_last_folder_path = os.path.join(merged_new_root, last_folder)

    os.makedirs(new_last_folder_path)

    for file in files:
        gen_merged_mid(os.path.join(root, file), os.path.join(new_last_folder_path, file))
```
