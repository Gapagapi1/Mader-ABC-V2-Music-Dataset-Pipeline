```python
import os
import music21
```

```python
new_root = "../../midi/lmd_matched_flat"
xml_root = "../../midi/lmd_matched_flat_musicxml_test"
```

```python
os.makedirs(xml_root)
```

```python
wrong_files = []
```

```python
for root, folders, files in os.walk(new_root):
    for file in files:
        midi_path = os.path.join(root, file)
        xml_folder = os.path.join(xml_root, root.replace(new_root + "/", ""))
        xml_path = os.path.join(xml_folder, file.replace(".midi", ".musicxml").replace(".mid", ".musicxml"))

        os.makedirs(xml_folder, exist_ok=True)
        
        try:
            score = music21.converter.parse(midi_path)
            score.write('musicxml', fp=xml_path)
        except Exception as e:
            wrong_files.append(midi_path)
            print(f"Error executing command on {midi_path}: {e}")
```

```python
wrong_files
```

```python
import json
with open("wrongs.json", "w") as wrongs_file:
    json.dump(wrong_files, wrongs_file)
```
