```python
import music21
```

```python
from music21 import converter

# Parse the MIDI file into a music21 stream
score = converter.parse('c24989559d170135b9c6546d1d2df20b.mid')

# Create an instance of the ABC converter
abc_converter = converter.subConverters.ConverterABC()

# Manually set the output extensions for ABC conversion
abc_converter.registerOutputExtensions = ('abc',)

# Convert the score to an ABC-formatted string, specifying the 'abc' format
abc_text = abc_converter.toData(score, 'abc')

# Write the ABC notation to a file
with open('c24989559d170135b9c6546d1d2df20b.abc', 'w') as f:
    f.write(abc_text)
```
