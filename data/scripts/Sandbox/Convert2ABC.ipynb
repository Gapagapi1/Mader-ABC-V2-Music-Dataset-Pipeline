```python
import os
import mido
import subprocess
```

```python
new_root = "../../midi/lmd_matched_flat_sanitized"
abc_root = "../../midi/lmd_matched_flat_abc_test5"
```

```python
os.makedirs(abc_root)
```

```python
wrong_files = []
```

```python
for root, folders, files in os.walk(new_root):
    for file in files:
        midi_path = os.path.join(root, file)
        abc_folder = os.path.join(abc_root, root.replace(new_root + "/", ""))
        abc_path = os.path.join(abc_folder, file.replace(".midi", ".abc").replace(".mid", ".abc"))
        
        # L:1/1 | -K:C | one bar per line | no group | no lyrics -aul 1 
        command = "midi2abc.exe -f {} -k 0 -obpl -nogr -noly -o {}".format(midi_path, abc_path)

        os.makedirs(abc_folder, exist_ok=True)
        
        try:
            subprocess.run(command, shell=True, check=True)
        except subprocess.CalledProcessError as e:
            wrong_files.append(midi_path)
            print(f"Error executing command on {midi_path}: {e}")
```

```python
wrong_files
```

```python
import json
with open("wrongs.json", "w") as wrongs_file:
    json.dump(wrong_files, wrongs_file)
```

```python
def get_channels(path):
    midi = mido.MidiFile(path)
    channels = set()

    for track in midi.tracks:
        for msg in track:
            if hasattr(msg, 'channel'):
                channels.add(msg.channel)
    
    return sorted(channels)
```

```python
get_channels("lmd_matched_flat/TRAAAGR128F425B14B/1d9d16a9da90c090809c153754823c2b.mid")
```

```python
def get_channel_instruments(path):
    midi = mido.MidiFile(path)
    channel_instruments = {}

    for track in midi.tracks:
        for msg in track:
            if msg.type == 'program_change':
                if hasattr(msg, 'channel'):
                    channel_instruments[msg.channel] = msg.program

    return {ch: prog for ch, prog in channel_instruments.items()}
```

```python
get_channel_instruments("lmd_matched_flat/TRAAAGR128F425B14B/1d9d16a9da90c090809c153754823c2b.mid")
```
