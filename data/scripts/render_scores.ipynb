```python
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
file_name = "../match_scores.json"

with open(file_name, 'r') as file:
    data = json.load(file)
```

```python
# Be careful multiple midi can be in same track

scores = {}

for variant_midi in data:
    for midi in data[variant_midi]:
        if midi not in scores:
            scores[midi] = data[variant_midi][midi]
        elif scores[midi] < data[variant_midi][midi]:
            scores[midi] = data[variant_midi][midi]
len(scores)
```

```python
data = scores.values()
split_range = 0.1

bins = np.arange(0, 1 + split_range, split_range)
counts, bins, patches = plt.hist(data, bins=bins, color='skyblue', edgecolor='black')

total = sum(counts)
for count, patch in zip(counts, patches):
    percentage = (count / total) * 100
    x = patch.get_x() + patch.get_width() / 2
    y = patch.get_height()
    plt.text(x, y, f'{percentage:.1f}%', ha='center', va='bottom')

plt.xlabel('Values')
plt.xticks(bins)
plt.ylabel('Frequency')
plt.show()
```
